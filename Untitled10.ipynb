```python
def get_db_schema(db_path="/content/drive/MyDrive/movie.db"):  # Update to use external DB
    """Retrieve database schema to improve SQL generation."""
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
    tables = cursor.fetchall()
    schema_info = ""

    for table in tables:
        table_name = table[0]
        cursor.execute(f"PRAGMA table_info({table_name})")
        columns = cursor.fetchall()
        schema_info += f"Table {table_name}: " + ", ".join([col[1] for col in columns]) + "\n"

    conn.close()
    return schema_info
```

```python
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer

def generate_sql_query(natural_language_query, model, tokenizer, db_schema):
    """Convert a natural language query into an SQL query using Phi-3 with schema awareness."""
    prompt = f"Database schema:\n{db_schema}\nConvert this to SQL: {natural_language_query}"

    # Generate without specifying max_length to avoid the error
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)  # Use model.device to ensure correct device

    # Disable caching to avoid the error and use num_beams (corrected from nums_beams)
    output = model.generate(**inputs, use_cache=False, max_new_tokens=50, num_beams=2)

    sql_query = tokenizer.decode(output[0], skip_special_tokens=True)
    return sql_query
```

```python
def validate_sql_query(sql_query, db_path="/content/drive/MyDrive/movie.db"):  # Update to use external DB
    """Validate the generated SQL query before execution."""
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        cursor.execute(f"EXPLAIN QUERY PLAN {sql_query}")
        conn.close()
        return True
    except Exception as e:
        return False, str(e)

def execute_sql_query(sql_query, db_path="/content/drive/MyDrive/movie.db"):  # Update to use external DB
    """Execute the generated SQL query on a SQLite database."""
    try:
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()
        cursor.execute(sql_query)
        results = cursor.fetchall()
        conn.close()
        return results
    except Exception as e:
        return f"Error: {e}"
```

```python
def ask_sql_query(user_query):
    """
    Converts a natural language question into an SQL query, validates, and executes it.
    Ensures the query is correctly formatted without extra text.
    """
    prompt = f"""
    Given the following database schema:
    {db_schema}

    Convert the following user request into a precise SQL query without extra explanations:
    "{user_query}"

    SQL Query:
    """

    # Move inputs to the same device as the model (model.device)
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(model.device)
    # Use max_new_tokens instead of max_length to control the generated output length
    # Disable caching by setting use_cache=False
    output = model.generate(**inputs, max_new_tokens=50, eos_token_id=tokenizer.eos_token_id, use_cache=False)
    sql_query = tokenizer.decode(output[0], skip_special_tokens=True).strip()

    # Ensure only SQL output is returned
    if "SQL Query:" in sql_query:
        sql_query = sql_query.split("SQL Query:")[-1].strip()

    print("\nGenerated SQL Query:", sql_query)

    # Validate SQL before execution
    is_valid, error_message = validate_sql_query(sql_query)
    if is_valid:
        results = execute_sql_query(sql_query)
        print("\nQuery Results:", results)
        return results
    else:
        print("\nInvalid SQL Query:", error_message)
        return None
```

```python
ask_sql_query("List the top 5 highest-grossing movies")
ask_sql_query("Find all movies released after 2010 with a rating above 7.5")
```
